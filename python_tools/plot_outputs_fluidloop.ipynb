```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

Opening of the files, put your working directory

```python
dscst = xr.open_dataset('file_path_Rz_cst')
```

```python
dsvar = xr.open_dataset('file_path_Rz_var')
```

Velocity

```python
dscst.w.plot()
plt.show()
```

Temperature

```python
dscst.theta.plot()
plt.title('Rz constant')
plt.show()
```

```python
dsvar.theta.plot()
plt.title('Rz varying')
plt.show()
```

Salinity

```python
dscst.salt.plot()
plt.title('Rz constant')
plt.show()
```

```python
dsvar.salt.plot()
plt.title('Rz varying')
plt.show()
```

Temperature evolution

```python
dsvar.theta[0].plot(label = '0')
plt.legend()
dsvar.theta[10].plot(label = '10')
plt.legend()
dsvar.theta[20].plot(label = '20')
plt.legend()
dsvar.theta[30].plot(label = '30')
plt.legend()
plt.title('Rz varying')

plt.show()
```

Salinity evolution

```python
dsvar.salt[0].plot(label = '0')
plt.legend()
dsvar.salt[10].plot(label = '10')
plt.legend()
dsvar.salt[20].plot(label = '20')
plt.legend()
dsvar.salt[30].plot(label = '30')
plt.legend()
plt.title('Rz varying')

plt.show()
```

Density evolution

```python
dsvar.sigma[0].plot(label = '0')
plt.legend()
dsvar.sigma[10].plot(label = '10')
plt.legend()
dsvar.sigma[20].plot(label = '20')
plt.legend()
dsvar.sigma[30].plot(label = '30')
plt.legend()
plt.title('Rz varying')
plt.show()
```
